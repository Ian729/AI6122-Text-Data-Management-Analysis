```python
import json,random
import pandas as pd
import spacy
import nltk
import dataframe_image as dfi
nlp = spacy.load("en_core_web_sm")

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data_set = []
#load yelp reviews.json
for line in open('yelp_academic_dataset_review.json', 'rb'):
    data_set.append(json.loads(line))
```

```python
#list to string method
def list_to_string(list_):
    str1 = "" 
    
    # traverse in the string  
    for ele in list_: 
        str1 += ele + " " 
    
    # return string  
    return str1 
```

```python
#randomly select 5 sentences
sents_5 =[]
while len(sents_5) < 5:
    ran = random.randint(0,len(data_set)-1)
    doc = nlp(data_set[ran]['text'])
    for sent in doc.sents:
        sents_5.append(sent.text)
        if len(sents_5) == 5:
            break
print(sents_5)
```

```python
#nltk
sents_5_onestring = list_to_string(sents_5)
text = nltk.word_tokenize(sents_5_onestring)
#words=[word.lower() for word in text if word.isalnum()]
result = nltk.pos_tag(text)

result_dict = dict(result)
print(len(result_dict))
df = pd.DataFrame.from_dict(result_dict,orient='index',columns=['NLTK'])
display(df)
```

```python
#spacy
doc = nlp(sents_5_onestring)
spacy_result={}
token_list=[]
for token in doc:
    #token.text.lower() not in token_list and
        spacy_result[token.text]= token.tag_
df_1 = pd.DataFrame.from_dict(spacy_result,orient='index',columns=['spaCy'])
print(len(spacy_result))
display(df_1)
```

```python
df['spaCy']=pd.Series(spacy_result.values(),index=spacy_result.keys())

display(df)
dfi.export(df,"3.2pos.png")
```
